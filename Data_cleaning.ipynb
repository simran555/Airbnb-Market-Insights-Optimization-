```python
def remove_outliers(df, column_name):
    # Calculate the IQR (Interquartile Range)
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    IQR = Q3 - Q1

    # Define the lower and upper bound for the outliers
    lower_bound = Q1 - (1.5 * IQR)
    upper_bound = Q3 + (1.5 * IQR)

    # Filter out the outliers
    df_filtered = df[
        (df[column_name] >= lower_bound) & 
        (df[column_name] <= upper_bound)
    ]

    return df_filtered, lower_bound, upper_bound
```

```python

def impute_skewed_data(df, column_name, impute_method=None):
    """
    Impute missing values in a DataFrame column based on the skewness of the data.
    
    Parameters:
        df (pd.DataFrame): DataFrame containing the data.
        column_name (str): Name of the column to check skewness and impute missing values.
        impute_method (str, optional): Specify 'mean' or 'median' to override automatic method. Default is None.
        
    Returns:
        None: The function modifies the DataFrame in place.
    """
    skewness = df[column_name].skew()
    print(f"Skewness: {skewness}")

    if impute_method is None:
        if skewness > 1 or skewness < -1:
            print("Highly skewed, using median for imputation.")
            impute_value = df[column_name].median()
        else:
            print("Not highly skewed, using mean for imputation.")
            impute_value = df[column_name].mean()
    elif impute_method == 'median':
        print("Forced to use median for imputation.")
        impute_value = df[column_name].median()
    elif impute_method == 'mean':
        print("Forced to use mean for imputation.")
        impute_value = df[column_name].mean()
    else:
        raise ValueError("Invalid imputation method. Use 'mean' or 'median'.")

    df[column_name].fillna(impute_value, inplace=True)

    
```
